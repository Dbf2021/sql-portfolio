```python
# Define the groups for A/B testing based on shipping type
group_a = df[df['shipping_type'] == 'Second Class']
group_b = df[df['shipping_type'] == 'Standard Class']

# Display group sizes
print(f"Second Class size: {len(group_a)}")
print(f"Standard Class size: {len(group_b)}")
```

```python
from scipy.stats import ttest_ind

# Perform t-test for sales per order
t_stat, p_value = ttest_ind(group_a['sales_per_order'], group_b['sales_per_order'])

print(f"T-Statistic: {t_stat}")
print(f"P-Value: {p_value}")

# Interpret results
if p_value < 0.05:
    print("Significant difference in sales per order between shipping types.")
else:
    print("No significant difference in sales per order between shipping types.")
```

# Perform a statistical test (e.g., t-test) to determine if there is a significant difference in sales and profit between the two groups.

```python
from scipy import stats

# Perform t-test for sales
sales_a = group_a['sales_per_order']
sales_b = group_b['sales_per_order']
sales_ttest = stats.ttest_ind(sales_a, sales_b)

# Perform t-test for profit
profit_a = group_a['profit_per_order']
profit_b = group_b['profit_per_order']
profit_ttest = stats.ttest_ind(profit_a, profit_b)

# Display t-test results
print(f"Sales t-test result: {sales_ttest}")
print(f"Profit t-test result: {profit_ttest}")
```

```python
import scipy.stats as stats

# Create a contingency table for 'delivery_status' and 'shipping_type'
contingency_table = pd.crosstab(df['delivery_status'], df['shipping_type'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Display the results
print(f"Chi-Square Statistic: {chi2_stat}")
print(f"P-Value: {p_value}")
print(f"Degrees of Freedom: {dof}")
print(f"Expected Frequencies Table:\n{expected}")

# Interpret the result
if p_value < 0.05:
    print("There is a significant relationship between delivery status and shipping type.")
else:
    print("No significant relationship between delivery status and shipping type.")
```

```python
from scipy.stats import shapiro

# Test for normality on the 'sales_per_order' column
stat, p_value_sales = shapiro(df['sales_per_order'].dropna())  # Drop NaN values before testing

# Display the results
print(f"Shapiro-Wilk Statistic: {stat}")
print(f"P-Value: {p_value_sales}")

# Interpret the result
if p_value_sales < 0.05:
    print("The 'sales_per_order' data is not normally distributed.")
else:
    print("The 'sales_per_order' data is normally distributed.")
```

### General Conclusion of the Case Study

This case study aimed to explore and analyze key performance indicators (KPIs) for an e-commerce company, focusing on aspects like sales per order, profit per order, shipment delays, and the influence of different shipping methods. The insights derived from this analysis can help the company optimize its business operations, improve customer satisfaction, and enhance profitability.

The following conclusions were drawn from the analysis:

1. **Sales and Profit Analysis**:
   - Both **sales per order** and **profit per order** are critical KPIs for understanding business performance. The data indicated that while most transactions involve average sales and profits, there are notable high-value outliers. Despite the presence of these outliers, the analysis found no significant difference between shipping methods (`Second Class` vs. `Standard Class`) in terms of their impact on sales or profit. This suggests that shipping type is not a key determinant of transaction value or profitability. Therefore, businesses should consider focusing on other factors such as pricing strategies, marketing, and customer segmentation to drive sales.

2. **Shipment Delay**:
   - **Shipment delay** was a significant factor affecting the customer experience. The analysis revealed that there were differences in shipment delays between the two shipping methods. A statistical t-test showed that the delays between `Second Class` and `Standard Class` shipping methods were significantly different, indicating that delivery speed could play a role in customer satisfaction. Customers who choose `Standard Class` may experience longer wait times, which could negatively impact their experience. As such, improving delivery times, particularly for slower shipping methods, should be a key focus for the company.

3. **Regional Performance**:
   - The **regional sales** and **profit** analysis showed variability in performance across different regions. Some regions generated higher sales and profits, suggesting that localized strategies could be beneficial. By focusing marketing efforts on high-performing regions and identifying potential in underperforming regions, the company can increase market penetration and optimize resource allocation.

4. **Data Integrity and Outlier Management**:
   - Outlier detection revealed a few high-value transactions that could skew the analysis. While these outliers may represent genuine high-value customers, they could also be data errors. Therefore, it's crucial to distinguish between legitimate high-value orders and anomalies that might distort the company's understanding of its performance.

5. **Strategic Recommendations**:
   - Based on the findings, it is recommended that the company focuses on improving **delivery speed** for slower shipping methods, particularly where delays are noticeable. Additionally, it could optimize marketing and sales strategies by focusing on high-performing regions and investigating the reasons behind low-performing regions.
   - The company should also consider enhancing **customer satisfaction** by addressing shipment delays and offering more tailored services. Analyzing **outliers** and understanding their nature (whether they are legitimate high-value transactions or data issues) could offer valuable insights into customer behavior and opportunities for growth.
   
In conclusion, the analysis has provided a clear view of the company's current performance, highlighting the need for operational improvements and strategic optimizations. By leveraging these insights, the company can make data-driven decisions to enhance its overall performance and better meet customer needs.
